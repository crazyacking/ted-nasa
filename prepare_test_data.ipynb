```python
# load RUL data
RUL_FD001 = pd.read_table("./CMAPSSData/RUL_FD001.txt",header=None,delim_whitespace=True)
RUL_FD002 = pd.read_table("./CMAPSSData/RUL_FD002.txt",header=None,delim_whitespace=True)
RUL_FD003 = pd.read_table("./CMAPSSData/RUL_FD003.txt",header=None,delim_whitespace=True)
RUL_FD004 = pd.read_table("./CMAPSSData/RUL_FD004.txt",header=None,delim_whitespace=True)
RUL_FD001.columns = ['RUL']
RUL_FD002.columns = ['RUL']
RUL_FD003.columns = ['RUL']
RUL_FD004.columns = ['RUL']
```

```python
def compute_rul_of_one_id(test_FD00X_of_one_id, max_cycle):
    '''
    输入train_FD001的一个engine_id的数据，输出这些数据对应的RUL（剩余寿命），type为list
    '''
    # 故障时的cycle, 由RUL_FD001给出
    max_cycle = max(test_FD00X_of_one_id['cycle']) + max_cycle
    rul_of_one_id = max_cycle - test_FD00X_of_one_id['cycle']
    return rul_of_one_id.tolist()

def compute_rul_of_one_file(test_FD00X,  RUL_FD00X):
    '''
    输入train_FD001，输出一个list
    '''
    rul = []
    # 循环train中，''engine_id''这一列的每一种id值
    for id in set(test_FD00X['engine_id']):
        rul.extend(compute_rul_of_one_id(test_FD00X[test_FD00X['engine_id'] == id], int(RUL_FD00X.iloc[ id - 1])))
    return rul
```

```python
def compute_scaled_rul_of_one_id(test_FD00X_of_one_id, max_cycle):
    '''
    输入train_FD001的一个engine_id的数据，输出这些数据对应的RUL（剩余寿命），type为list
    '''
    # 故障时的cycle, 由RUL_FD001给出
    max_cycle = max(test_FD00X_of_one_id['cycle']) + max_cycle
    rul_of_one_id = max_cycle - test_FD00X_of_one_id['cycle']
    scaled_rul_of_one_id = rul_of_one_id / max_cycle
    return scaled_rul_of_one_id.tolist()

def compute_scaled_rul_of_one_file(test_FD00X,  RUL_FD00X):
    '''
    输入train_FD001，输出一个list
    '''
    rul = []
    # 循环train中，''engine_id''这一列的每一种id值
    for id in set(test_FD00X['engine_id']):
        rul.extend(compute_scaled_rul_of_one_id(test_FD00X[test_FD00X['engine_id'] == id], int(RUL_FD00X.iloc[ id - 1])))
    return rul
```

```python
# 为4个data增加RUL列
for (data_file, rul_file) in [('test_FD00' + str(i), 'RUL_FD00' + str(i)) for  i in range(1,5)]:
    # have a look at the info of each data file
    eval(data_file)['RUL'] = compute_scaled_rul_of_one_file(eval(data_file), eval(rul_file))
```
